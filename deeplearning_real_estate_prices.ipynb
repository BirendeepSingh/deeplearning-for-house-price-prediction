```python
# Convert the date column to datetime type
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Drop rows with NaT values in the date column
df = df.dropna(subset=['date'])

f, ax = plt.subplots(figsize=(20,20))
sns.heatmap(df.corr(), annot=True)
```

```python
df_sample = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'yr_built']]
df_sample.head()
```

```python
sns.pairplot(df_sample)
```

```python
selected_features = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'yr_built']
x=df_sample[selected_features]
```

```python
y=df_sample['price']
```

```python
x.shape, y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
scaler.data_max_
```

```python
scaler.data_min_
```

```python
y=y.values.reshape(-1,1)
```

```python
y_scaled=scaler.fit_transform(y)
y_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf

import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(100,input_dim=8,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(200,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
epochs_hist=model.fit(x_train,y_train,epochs=100,batch_size=50,validation_split=0.2)
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.ylabel('Training and Validation Loss')
plt.xlabel('Epoch Number')
plt.legend(['Training Loss', 'Validation Loss'])
```

```python
# Ensure the input array has 8 features
x_test1 = np.array([4, 3, 1960, 5000,1, 2000, 3000, 1960]).reshape(1, -1)
scaler_1 = MinMaxScaler()
x_test1_scaled = scaler_1.fit_transform(x_test1)
y_predict1 = model.predict(x_test1_scaled)
y_predict1 = scaler.inverse_transform(y_predict1)
y_predict1
```

```python
y_predict=model.predict(x_test)
plt.plot(y_test,y_predict,'^',color='r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
y_predict_orig = scaler.inverse_transform(y_predict)
y_test_orig = scaler.inverse_transform(y_test)
```

```python
plt.plot(y_test_orig, y_predict_orig, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
plt.xlim(0, 5000000)
plt.ylim(0, 3000000)
```

```python
k = x_test.shape[1]
n = len(x_test)
n
```

```python
k
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
from math import sqrt
RMSE = float(format(np.sqrt(mean_squared_error(y_test_orig, y_predict_orig)),'.3f'))
MSE = mean_squared_error(y_test_orig, y_predict_orig)
MAE = mean_absolute_error(y_test_orig, y_predict_orig)
r2 = r2_score(y_test_orig, y_predict_orig)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2)
```

```python
from tensorflow.keras.layers import Dropout

model.add(Dense(20, input_dim=20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
epochs_hist=model.fit(x_train,y_train,epochs=100,batch_size=50,validation_split=0.2)
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.ylabel('Training and Validation Loss')
plt.xlabel('Epoch Number')
plt.legend(['Training Loss', 'Validation Loss'])
```

```python
# Ensure the model is defined and trained before running this cell
y_predict = model.predict(x_test)
plt.plot(y_test, y_predict, "^", color='r')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.show()
```
